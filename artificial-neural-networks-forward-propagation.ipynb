```python
import numpy as np

#initializing weights and biases
weights = np.around(np.random.uniform(size=6), decimals=2)
biases = np.around(np.random.uniform(size=3), decimals=2)
```

```python
print(weights)
print(biases)
```

### computing output for given inputs

```python
x_1 = 0.5
x_2 = 0.85
print('x1 is {} and x2 is {}'.format(x_1, x_2))

# compute the weighted sum in the first node of hidden layer, z_11
z_11 = x_1 * weights[0] + x_2 * weights[1] + biases[0]

print('The weighted sum of the inputs at z_11 {}'.format(z_11))

# compute the weighted sum in the first node of hidden layer, z_12
z_12 = x_1 * weights[2] + x_2 * weights[3] + biases[1]

print('The weighted sum of the inputs at z_12 {}'.format(z_12))
```

```python
# sigmoid of z_11
a_11 = 1.0 / (1.0 + np.exp(-z_11))
a_12 = 1.0 / (1.0 + np.exp(-z_12))

print('The activation of the first node a_11 is {}'.format(np.around(a_11, decimals=4)))
print('The activation of the second node a_12 is {}'.format(np.around(a_12, decimals=4)))
```

```python
# calculate output layer weighted sum
z_2 = a_11 * weights[4] + a_12 * weights[5] + biases[2]
print('The weighted sum of the inputs at the single node in the output layer is {}'.format(np.around(z_2, decimals=4)))
```

```python
# calculate output sigmoid
a_2 = 1.0 / (1.0 + np.exp(-z_2))
print('The output of the network for x1 = 0.5 and x2 = 0.85 is {}'.format(np.around(a_2, decimals=4)))
```

### Build a Neural Network

```python
# number of inputs
n = 2
# number of hidden layers
num_hidden_layers = 2 
# number of nodes in each hidden layer
m = [2, 2] 
# number of nodes in the output layer
num_nodes_output = 1 
```

#### initialize weights

```python

# import the Numpy library
import numpy as np 

# number of nodes in the previous layer
num_nodes_previous = n 

# initialize network an an empty dictionary
network = {} 

# loop through each layer and randomly initialize the weights and biases associated with each node
# notice how we are adding 1 to the number of hidden layers in order to include the output layer
for layer in range(num_hidden_layers + 1): 
    
    # determine name of layer
    if layer == num_hidden_layers:
        layer_name = 'output'
        num_nodes = num_nodes_output
    else:
        layer_name = 'layer_{}'.format(layer + 1)
        num_nodes = m[layer]
    
    # initialize weights and biases associated with each node in the current layer
    network[layer_name] = {}
    for node in range(num_nodes):
        node_name = 'node_{}'.format(node+1)
        network[layer_name][node_name] = {
            'weights': np.around(np.random.uniform(size=num_nodes_previous), decimals=2),
            'bias': np.around(np.random.uniform(size=1), decimals=2),
        }
    
    num_nodes_previous = num_nodes
    
print(network) # print network
```
